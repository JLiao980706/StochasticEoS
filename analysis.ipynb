## Load Predicted Dynamics

```python
import torch
import numpy as np

PRED_PATH = '/content/drive/My Drive/Research/StochasticEoS/predicted_dynamics.pt'

data = torch.load(PRED_PATH, map_location='cpu')

# Each entry is a list of num_runs lists, each of length num_steps.
# Convert to numpy arrays of shape (num_runs, num_steps) for convenience.
x_hat      = np.array(data['x_hat'])       # (num_runs, num_steps)  <v_hat_t, u_t>
y_hat      = np.array(data['y_hat'])       # (num_runs, num_steps)  <nabla_S_t^perp, v_hat_t>
alpha      = np.array(data['alpha'])       # (num_runs, num_steps)  -<nabla_S_t, nabla_L_t>
beta       = np.array(data['beta'])        # (num_runs, num_steps)  ||nabla_S_t^perp||^2
noise_u    = np.array(data['noise_u'])     # (num_runs, num_steps)  <xi_t, u_t>
noise_sharp = np.array(data['noise_sharp'])# (num_runs, num_steps)  <xi_t, nabla_S_t>

num_runs, num_steps = x_hat.shape
print(f'Loaded {num_runs} runs, {num_steps} steps each.')
print(f'x_hat      shape: {x_hat.shape}')
print(f'y_hat      shape: {y_hat.shape}')
print(f'alpha      shape: {alpha.shape}')
print(f'beta       shape: {beta.shape}')
print(f'noise_u    shape: {noise_u.shape}')
print(f'noise_sharp shape: {noise_sharp.shape}')
```
